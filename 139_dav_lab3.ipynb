```python
#1


def create_hierarchical_index(df):
  if 'Product' in df.columns and 'Month' in df.columns:
    print("\n Hierarchical index created suceesfully")
    df_hierarchical=df.set_index(['Product', 'Month'])
    return df_hierarchical
  else:
    print("the required columns month and product are not present")
    return df
hierarchical_sales=create_hierarchical_index(sales_df)
print(f"\n{hierarchical_sales}")
```

```python
if isinstance(hierarchical_sales.index,pd.MultiIndex):
  print("Hierarchical index Example-Subsetting:")
  try:
    user_product= input("Enter a product name to view its sales: ")
    product_sales= hierarchical_sales.xs(user_product, level='Product')
    print(f"Sales for {user_product}:\n{product_sales.head()}")
  except KeyError:
    print("Product {user_product} not found in the data . Plkease enter a valid product name")
  except Exception as e:
    print(f"An error occured: {e}")
```

```python

def demonstrate_merges(sales_df,feedback_df):
  print("\n Merge operation:\n")

  #inner_join
  inner_merged=pd.merge(sales_df,feedback_df,on="OrderID",how='inner')
  print(f"inner join : {inner_merged.shape} rows")
  print(f"inner merged table{inner_merged}")
  print("Use case: when you only want orders have feedback")
  print("\n")

  #left join
  left_merged = pd.merge(sales_df, feedback_df, on="OrderID", how='left')
  print(f"Left Join: {left_merged.shape} rows")
  print(f"left merged table{left_merged}")
  print("Use case: When you want all orders, including those without feedback")
  print("\n")

  # RIGHT JOIN
  right_merged = pd.merge(sales_df, feedback_df, on="OrderID", how='right')
  print(f"Right Join: {right_merged.shape} rows")
  print(f"right merged table{right_merged}")
  print("Use case: When you want all feedback entries, including those without orders (rare case)")
  print("\n")

  # OUTER JOIN (Full outer join)
  outer_merged = pd.merge(sales_df, feedback_df, on="OrderID", how='outer')
  print(f"Outer Join: {outer_merged.shape} rows")
  print(f"outer merged table{outer_merged}")
  print("Use case: When you want all orders and all feedback, matched where possible")
  print("\n")



demonstrate_merges(sales_df,feedback_df)
```

```python
#3
def demonstrate_concatetation(sales_df,feedback_df):
  print("\n Concatenation operation:\n")

  q1=sales_df.iloc[:1].copy()
  q2=sales_df.iloc[1:3].copy()
  q3=sales_df.iloc[3:].copy()


  print("Quartile 1:\n",q1)
  print("Quartile 2:\n",q2)
  print("Quartile 3:\n",q3)


  q1['Quarter'] = 'Q1'
  q2['Quarter'] = 'Q2'
  q3['Quarter'] = 'Q3'


  vertical_concatenated = pd.concat([q1, q2, q3], axis=0)
  print(f"\nVertical Concatenation:\n{vertical_concatenated.shape} rows")
  print("Implications: Combines similar data from different time periods or batches.")
  print("Use case: When you want to append data from multiple sources with the same structure.\n")
  print(vertical_concatenated)


  sales_metrics=sales_df[['OrderID','Sales']].set_index('OrderID')
  product_metrics=sales_df[['OrderID','Product']].set_index('OrderID')
  print("\nSales Metrics:\n",sales_metrics)
  print("\nProduct Metrics:\n",product_metrics)



  horizontal_concatenated = pd.concat([sales_metrics, product_metrics], axis=1)
  print(f"\nHorizontal Concatenation:\n{horizontal_concatenated.shape} columns")
  print("Implications: Combines different features/attributes for the same set of rows.")
  print("Use case: When you want to merge datasets with different variables but related rows.\n")
  print(horizontal_concatenated)




demonstrate_concatetation(sales_df,feedback_df)
```

```python
#4

import pandas as pd

def combine_with_feedback_fill(sales_df, feedback_df):
    print("\nCombining Data with Missing Value Filling:")
    combined_df = pd.merge(sales_df, feedback_df, on='OrderID', how='left')
    if 'Feedback_Score' in combined_df.columns:
        combined_df['Feedback_Score'] = combined_df['Feedback_Score'].fillna(8)
        print("Missing Feedback_Score values filled with 8")
    else:
        print("Feedback_Score column not found in merged data")
    return combined_df
combine_with_feedback_fill(sales_df,feedback_df)
```

```python
#5
def create_sales_pivot(sales_df):
    print("\nPivot Table Creation:")
    required_cols = ['Product', 'Month', 'Sales']
    try:
        if all(col in sales_df.columns for col in required_cols):
            pivot_df = sales_df.pivot_table(index='Product', columns='Month', values='Sales', fill_value=0)
            print("Pivot Table created successfully!")

            print(f"\nPivot Table Shape: {pivot_df.shape}")
            print("\nPivot Table:")
            print(pivot_df)

            print("\n" + "=" * 50)
            print("SALES SUMMARY")
            print("=" * 50)

            print("\nTotal Sales by Product:")
            product_totals = pivot_df.sum(axis=1)
            for product, total in product_totals.items():
                print(f"{product}: Rs. {total:,.2f}")

            print("\nTotal Sales by Month:")
            month_totals = pivot_df.sum(axis=0)
            for month, total in month_totals.items():
                print(f"{month}: Rs. {total:,.2f}")

            grand_total = pivot_df.sum().sum()
            print("-" * 50)
            print(f"\nGrand Total Sales: Rs. {grand_total:,.2f}")
            print("-" * 50)

            return pivot_df
        else:
            missing = [col for col in required_cols if col not in sales_df.columns]
            print(f"Missing required columns: {missing}")
            return None
    except Exception as e:
        print(f"Error creating pivot table: {e}")
        return None



create_sales_pivot(sales_df)
print("\nANALYSIS COMPLETE")
```
