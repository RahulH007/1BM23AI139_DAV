```python
import pandas as pd
```

```python
df=pd.read_csv('Sales_data.csv')
```

```python
df
```

```python
# 1. Pivot the data to calculate the total revenue generated by each salesperson on each date.
pivot_revenue_date_salesperson = df.pivot_table(index='Date', columns='Salesperson', values='Revenue', aggfunc='sum')
print("Total revenue generated by each salesperson on each date:")
display(pivot_revenue_date_salesperson)
```

```python
# 2. Find the average revenue per sale for each product.
average_revenue_per_product = df.groupby('Product')['Revenue'].mean()
print("\nAverage revenue per sale for each product:")
display(average_revenue_per_product)
```

```python
# 3. Find the maximum number of units sold in a single transaction by each salesperson.
max_units_sold_per_salesperson = df.groupby('Salesperson')['Units Sold'].max()
print("\nMaximum number of units sold in a single transaction by each salesperson:")
display(max_units_sold_per_salesperson)
```

```python
# 4. Show the percentage of total revenue contributed by each region.
total_revenue = df['Revenue'].sum()
revenue_by_region = df.groupby('Region')['Revenue'].sum()
percentage_revenue_by_region = (revenue_by_region / total_revenue) * 100
print("\nPercentage of total revenue contributed by each region:")
display(percentage_revenue_by_region)
```

```python
# 5. Determine which salesperson completed the most sales transactions (count of unique transactions)
sales_transactions_count = df.groupby('Salesperson').size().sort_values(ascending=False)
most_sales_salesperson = sales_transactions_count.index[0]
print(f"\nSalesperson with the most sales transactions: {most_sales_salesperson}")
display(sales_transactions_count)
```

```python
# 6. Pivot the data to show both the total revenue and total units sold by each salesperson for each product.
pivot_revenue_units_salesperson_product = df.pivot_table(index='Salesperson', columns='Product', values=['Revenue', 'Units Sold'], aggfunc='sum')
print("\nTotal revenue and total units sold by each salesperson for each product:")
display(pivot_revenue_units_salesperson_product)
```

```python
# 7. Create a pivot table to display the total units sold in each region on each date.
pivot_units_region_date = df.pivot_table(index='Date', columns='Region', values='Units Sold', aggfunc='sum')
print("\nTotal units sold in each region on each date:")
display(pivot_units_region_date)
```
