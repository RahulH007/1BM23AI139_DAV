```python
import pandas as pd

df=pd.read_csv('/DtaFrame1.csv')
print(df.head(5))
```

```python
#prin ting the null values
missing_occ=df['Occupation'].isnull()
missing_sat=df['Satisfaction_Level'].isnull()
missing_occ
missing_sat
print("the missing occupation is",missing_occ.sum())
print("the missing satisfaction level is",missing_sat.sum())
```

```python
filled_df=df.fillna(0)
print(filled_df)


dropped_df=df.dropna()
print(dropped_df)
```

```python
df['satisfaction_level_binary']=df['Satisfaction_Level'].apply(lambda x:'high' if x>0.5 else 'low')
df
```

```python
purchase_map={ 'low':0 , 'medium': 1,'high':2}
df['purchase numeric']=df['Purchase_History'].map(purchase_map)
df
```

```python
#iqr
q1=df['Income'].quantile(0.25)

q3=df['Income'].quantile(0.75)

lower_bound=q1-1.5*(q3-q1)
upper_bound=q3+1.5*(q3-q1)


rr=df[(df['Income']>=lower_bound) & (df['Income']<=upper_bound)]
rr
```

```python
yy=df[(df['Income']<lower_bound) | (df['Income']>upper_bound)]
yy
```

```python
missing_years_employed=df['Years_Employed'].isnull()
missing_years_employed
print("the missing years employed is",missing_years_employed.sum())

df['Years_Employed'].fillna(df['Years_Employed'].mean(),inplace=True)
df
```
