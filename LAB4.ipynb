```python
import numpy as np

# Using Pandas functions
print("Pandas summary statistics:")
print("Population mean (Pandas):", df['Population'].mean())
print("GDP sum (Pandas):", df['GDP'].sum())
print("Population max (Pandas):", df['Population'].max())
print("GDP min (Pandas):", df['GDP'].min())

print("\nNumPy summary statistics:")
# Using NumPy functions
print("Population mean (NumPy):", np.mean(df['Population']))
print("GDP standard deviation (NumPy):", np.std(df['GDP']))
print("Population sum (NumPy):", np.sum(df['Population']))
```
