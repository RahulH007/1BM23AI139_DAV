```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset_9.csv")

plt.figure(figsize=(8,6))
plt.scatter(df["GrLivArea"], df["SalePrice"])
plt.title("GrLivArea vs SalePrice")
plt.xlabel("Above Ground Living Area (GrLivArea)")
plt.ylabel("Sale Price")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Dataset_9.csv")

cols = ["GrLivArea", "OverallQual", "TotalBsmtSF", "GarageCars",
        "GarageArea", "1stFlrSF", "YearBuilt", "SalePrice"]

corr = df[cols].corr()

plt.figure(figsize=(10,7))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Key Numerical Features")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset_9.csv")
plt.figure(figsize=(10,7))

plt.scatter(
    df["GrLivArea"],
    df["SalePrice"],
    s=df["OverallQual"] * 10,
    c=df["YearBuilt"],
    cmap="viridis",
    alpha=0.6
)

plt.colorbar(label="Year Built")
plt.title("Bubble Chart: GrLivArea vs SalePrice (Bubble = OverallQual)")
plt.xlabel("Above Ground Living Area (GrLivArea)")
plt.ylabel("Sale Price")
plt.grid(True)
plt.show()
```
