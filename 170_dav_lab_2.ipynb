```python
def missing_values(df):
  df['Age'].fillna(df['Age'].mean(),inplace=True)
  df['City'].fillna('unkown',inplace=True)
  return df
missing_values(df)
print("the null values are",df.isnull().sum())
df
```

```python
print("no of rows in datafrrame",len(df))
def remove_duplicates(df):
  df=df.copy()
  df=df.drop_duplicates()
  return df
df=remove_duplicates(df)
print("AFter deleting the duplicates")
df
```

```python
def replace_gender(df):
  replace={"M":"Male","F":"Female"}
  df['Gender'].replace(replace,inplace=True)
  return df
df=replace_gender(df)
df
```

```python
import numpy as np
def group_age(df):
  df=df.copy()
  grp={'0-17','18-30','30-40','40-50','50 +'}
  bins=[0,18,30,40,50,np.inf]
  df['Grp_Age']=pd.cut(df['Age'],bins=bins,labels=grp)
  return df
df=group_age(df)
print("dataframe after age grouping")
df
```

```python
def dummy_column(df):
  df=df.copy()
  if 'City' in df.columns:
    dummies=pd.get_dummies(df['City'],prefix='City')
    df=pd.concat([df,dummies],axis=1)
  return df
df=dummy_column(df)
print("dataframe after converting city to dummy variables")
df
```
